```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import normalize
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
```

```python
img_org = Image.open('test_pca1.jpeg').convert('L')
plt.imshow(img_org)
plt.show()
```

```python
img_org.size
```

```python
pca = PCA(n_components=25) #(.99)
img_norm = normalize(img_org)
lower_dimension_data = pca.fit_transform(img_norm)
print(lower_dimension_data.shape)
print(pca.explained_variance_ratio_)
print('Variance retained: ', ((1 - sum(pca.explained_variance_ratio_) / len(pca.explained_variance_ratio_)) * 100))
```

```python
approximation = pca.inverse_transform(lower_dimension_data)
approximation = approximation * 255

_, (ax1, ax2)  = plt.subplots(1,2,figsize=(15,8))
ax1.imshow(img_org)
ax2.imshow(approximation, cmap='gray')
plt.show()
```
