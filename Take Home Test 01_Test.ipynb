```python
# Missing values
def mis_data(df):
    total = df.isnull().sum().sort_values(ascending = False)
    percent = round(total/len(df)*100,2)
    return pd.concat([total,percent], axis = 1, keys = ['Total','Percent'])
```

```python
mis_values = mis_data(data_train)
mis_values
```
